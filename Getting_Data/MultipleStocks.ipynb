```python
import datetime as dt
import yfinance as yf
import pandas as pd
```

```python
stocks= ["AMZN","MSFT", "INTC", "GOOG", "INFY.NS", "3988.HK"]
start = dt.datetime.today()-datetime.timedelta(360)
end   = dt.datetime.today()

c1_price = pd.DataFrame() # empty dataframe which will be filled with closing prices
ohlcv_data = {}
```

```python
# Looping over tickers and creating a dataframe with close prices
for ticker in stocks:
    c1_price[ticker] =  yf.download(ticker,start,end)["Adj Close"]
```
