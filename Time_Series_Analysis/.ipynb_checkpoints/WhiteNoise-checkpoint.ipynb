# White Noise

```r
set.seed(1)
x <- rnorm(1000, mean=0, sd=1)
plot(x)
acf(x)
# The first value, acf(0)=1 the series correlates with itself
# the autocorrelation is approximately 0: these values are independent
# mu_w = E[w_t] = 0 (mean of a white noise process is zero)
# Delta_k = Core(w_t, w_(t+k)) = 1 when k = 0
#                              = 0 when k=1,2,3,..
```

```r
# there is a trend: the acf() is approximately linear
x <- seq(1,1000)
plot(x)
acf(x)
```
