# Random Walk

```r
set.seed(2)
```

```r
# first we generate 100 white noise values
w <- rnorm(2000, mean=0, sd=1)
```

```r
# we create 100 items with value 0 for the x(t) time series
x <- rep(0,2000)
```

```r
# random walk is x(t) = x(t-1) (previouvalue) +w (white noise)
for(t in 2:2000) x[t] <- x[t-1] + w[t]
```

```r
# lets plot the random walk with lines ("l")
plot(x, type="l")
```

```r
# autocorrelation plot
acf(x)
```

```r
# Random walk is not a stationary process but we can make
# it stationary by using differencing transformation
acf(diff(x), na.action=na.omit)
# diff operator calculates x(t) - x(t-1) and creates a time series out of it
```
